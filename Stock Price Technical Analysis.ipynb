```python
stock = pd.DataFrame()
```

```python
stock['Close_msft'] = df.Close.msft
```

```python
import numpy as np
stock['returns_msft'] = np.log(df.Close.msft.div(df.Close.msft.shift(1)))
```

```python
type(stock)
```

```python
stock.drop(columns = ['prev_close'],inplace=True)
```

```python
stock.dropna(inplace=True)
```

```python
stock.returns_msft
```

# A simple Buy and hold strategy

```python
stock.returns_msft.hist(bins=100,figsize = (12,8))
```

```python
stock.returns_msft.sum()
```

```python
np.exp(stock.returns_msft.sum())
```

```python
stock['cumreturns_msft']= stock.returns_msft.cumsum().apply(np.exp)
```

```python
stock.cumreturns_msft.plot(figsize=(12,8))
```
